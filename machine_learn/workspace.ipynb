```python
import matplotlib.pyplot as plt

# ====  ====
models = [
    "Linear",
    "LSTM separate,absolute",
    "LSTM all,absolute",
    "LSTM separate,relative",
    "LSTM all,relative",
    "LSTM PVA"
]

ADE = [0.1055, 0.0732, 0.0647, 0.0626, 0.0654,0.0536]
FDE = [0.2106, 0.1368, 0.1172, 0.1216, 0.1197,0.0972]

# ==== ADEFDE====
x = range(len(models))
width = 0.35

fig, ax = plt.subplots(figsize=(9, 5))
ax.bar([i - width/2 for i in x], ADE, width, label="ADE")
ax.bar([i + width/2 for i in x], FDE, width, label="FDE")

ax.set_ylabel("Error")
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=20, ha='right')
ax.set_title("ADE/FDE")
ax.legend()
ax.grid(axis="y", linestyle="--", alpha=0.5)

# 
for i, (a, f) in enumerate(zip(ADE, FDE)):
    ax.text(i - width/2, a + 0.002, f"{a:.3f}", ha="center", va="bottom", fontsize=9)
    ax.text(i + width/2, f + 0.002, f"{f:.3f}", ha="center", va="bottom", fontsize=9)

plt.tight_layout()
plt.savefig("result2.png")
plt.show()

```
