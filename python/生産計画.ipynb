```python


import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

csv_path = Path(".csv")          
df = pd.read_csv(csv_path)
df.columns = df.columns.str.strip().str.lower()
x = df["start_rate"]           
y1 = df["outrate_parts"]       
y2 = df["mean_cycle"]          
fig, ax_left = plt.subplots(figsize=(8, 5))


ax_left.plot(x, y1, marker="o", linestyle="-", label="Outrate Parts", color="tab:blue")
ax_left.set_xlabel("Start Rate")
ax_left.set_ylabel("Outrate Parts", color="tab:blue")
ax_left.tick_params(axis="y", labelcolor="tab:blue")


ax_right = ax_left.twinx()
ax_right.plot(x, y2, marker="s", linestyle="--", label="Mean Cycle", color="tab:red")
ax_right.set_ylabel("Mean Cycle", color="tab:red")
ax_right.tick_params(axis="y", labelcolor="tab:red")

fig.legend(loc="upper left", bbox_to_anchor=(0.1, 1.0))
fig.tight_layout()

plt.savefig("result.png")
```
