従業員はパートタイムと正社員に分類され、また、正社員は管理職と一
般職に分類されるとする。パートタイム社員の残業はなしで、一般社員
は１０時間、また、管理職社員は一般社員の２倍であるとする。このクラ
ス構造を定義し、パート社員A、一般社員B、管理職社員Cのインスタンス
を生成し、その名前、職位、残業可能時間を出力するプログラムを作成
しなさい。

```python
class Employee:
    def __init__(self,name,position,overtime):
        self.name = name
        self.overtime = overtime
        self.position = position
    def info(self):
        print(f"名前: {self.name}, 職位: {self.position}, 残業可能時間: {self.overtime}時間")
class PartTime(Employee):
    def __init__(self, name):
        super().__init__(name, "パートタイム", 0)


class FullTime(Employee):
    def __init__(self, name):
        super().__init__(name, "一般職", 10)


class Manager(Employee):
    def __init__(self, name):
        super().__init__(name, "管理職", 20)
        
# インスタンス生成
a = PartTime("A")
b = FullTime("B")
c = Manager("C")

# 出力
a.info()
b.info()
c.info()
```

端末から入力する成績のGPAを計算するプログラムを変更し、成績’A+’,
…, ‘D’以外の文字を入力したら例外を発生し、”Invalid Input”を出力して
終了するように書きなおしなさい。
score = {'A+': 4.0, 'A': 3.0, 'B': 2.0, 'C': 1.0, 'D': 0.0}
ここで、
GPA = (4.0*(‘A+’の科目数)+3.0*(‘A’の科目数）+...)/科目総数

```python
def cal_gpa():
    score = {"A+":4.0,"A":3.0,"B":2.0,"C":1.0,"D":0.0}
    grades = []
    while True:
        s = input()
        if s == "":
            break
        try:
            sc = score[s]
            grades.append(sc)
        except KeyError:
            print("Invalid Input")
            return
    # 入力が1つ以上あれば平均（GPA）を計算
    if grades:
        gpa = sum(grades) / len(grades)
        print(f"Your GPA is {gpa:.2f}")  # 小数点以下2桁で表示
    else:
        print("No grades entered.")  # 空のまま終了した場合の対応
        
cal_gpa()
```

スタック構造を使って端末から入力した内容を逆順
にして出力するプログラムを作成しなさい。ただし、
スタックが空の場合、データを取得すると例外を発
生することを考えること。

```python
class Stack:
    def __init__(self):
        self._data = []
    def __len__(self):
        return self._data.__len__()
    def push(self,data):
        self._data.append(data)
    def is_empty(self):
        return len(self._data) == 0
    def pop(self):
        if self.is_empty():
            raise ValueError("Stack Is Empty")
        return self._data.pop()

s = Stack()
text = input()
# 1文字ずつスタックに積む
for ch in text:
    s.push(ch)

# スタックから取り出して逆順出力
reversed_text = ""
while not s.is_empty():
    reversed_text += s.pop()

print(reversed_text)
```

与えられた文字列に文字“(“, ”{“, ”[“, ”<“のいずれかと
ペアとなる文字”>”, “)”, “}”, “]”を検索し、ペアが見つ
かった場合はTrue、見つからない場合はFalseを出力す
るプログラムを作成し、下の文字列を使って出力結果を
示しなさい。

```python
text = input()

flag = True
for ch in text:
    if ch == '(' and text.find(')') == -1:
        flag = False
    if ch == '{' and text.find('}') == -1:
        flag = False
    if ch == '[' and text.find(']') == -1:
        flag = False
    if ch == '<' and text.find('>') == -1:
        flag = False
print(flag)
```

資料にある「指定容量をもつ行列」プログラムを参考に
して、リストの組み込み関数pop()およびappend()を
使って、行列のプログラムを書きなおし、実行しなさい。

```python
class Queue:
    def __init__(self):
        self._data = []

    def __len__(self):
        return len(self._data)

    def is_empty(self):
        return len(self._data) == 0

    def enqueue(self, obj):
        #データを末尾に追加
        self._data.append(obj)

    def dequeue(self):
        #先頭からデータを取り出す
        if self.is_empty():
            raise ValueError("Queue is empty")
        return self._data.pop(0)  # pop(0)で先頭を取り出す


# --- 動作確認 ---
q = Queue()
q.enqueue("A")
q.enqueue("B")
q.enqueue("C")

print(q.dequeue())  # → A
print(q.dequeue())  # → B
print(q.dequeue())  # → C

# 空のキューから取り出すと例外
# print(q.dequeue())  # ValueError: Queue is empty
```
