1. ３つの正の整数a, b, cを調べ、最も大きい奇数を表示するプログラムを作成
し、適当な数値例で実行して下さい。奇数がない場合、“No odd number in a,
b, c”を表示しなさい。

```python
def find_largest_odd(a, b, c):
    # 入力された3つの数値をリストに格納
    numbers = []
    numbers.append(a)
    numbers.append(b)
    numbers.append(c)

    # 最大の奇数を記録する変数。初期値は存在しないことを示す -1
    mx = -1

    # 各数値を順に確認
    for i in numbers:
        # 奇数なら、現在の最大値と比較して更新
        if i % 2 == 1:
            mx = max(mx, i)

    # 奇数が存在しなかった場合
    if mx == -1:
        print("No Odd number in a, b, c")
    # 奇数が存在した場合、その最大値を出力
    else:
        print(mx)
```

```python
find_largest_odd(1,2,3)

find_largest_odd(2,4,6)
```

例えば、2021年5月13日のように、YYYY年MM月DD日が与えられたら、その日
はその年の何日目になるかを求めるプログラムを作成し、2024年10月15日は
その年の何日目を示しなさい。ただし、閏年は
(1)４で割り切れる
(2)例外として、100で割り切れ400で割り切れなければ平年である。
注）PythonのDatetimeモジュールを使用しないこと。

```python
def cal_days(y, m, d):
    # 各月の日数を格納したリスト（平年）
    months = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    # うるう年判定：
    # 400で割り切れる年、または4で割り切れて100で割り切れない年はうるう年
    if (y % 400 == 0) or (y % 4 == 0 and y % 100 != 0):
        months[1] += 1  # 2月の日数を29日にする

    # その年の「m月d日」が年初から何日目かを求める
    res = d  # まず、今月の日付分を足す
    for i in range(m - 1):
        res += months[i]  # 1月から前月までの日数を合計

    return res  # 計算結果を返す
```

```python
cal_days(2024,10,15)
```

xの平方根を値low＋highの間で探すプログラムを完成し、実行しなさい。

```python
def cal_sqrt(x, numGuess):
    epsilon = 0.001
    low = 0
    high = max(1, x) 
    ans = (high + low) / 2
    while abs(ans**2 - x) >= epsilon:
        # 推定値の二乗がxより大きい → 答えは左側にある
        if ans**2 > x:
            high = ans
        # 推定値の二乗がxより小さい → 答えは右側にある
        else:
            low = ans
        ans = (high + low) / 2
    #print("GuessNum:", numGuess, "guess answer root of", x, "is:", ans)
```
