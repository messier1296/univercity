昇順チェーンを作るプログラムにあるオブジェクトobjを追加する関数
insert_item(self, obj) とチェーンの先頭データを削除する関数
delete_item(self) を完成し、実行結果を示しなさい。

```python
class sort_list:
    class _Node:
        def __init__(self, e, next):
            self._item = e
            self._next = next

    def __init__(self):
        self._head = None
        self._size = 0

    def is_empty(self):
        return self._size == 0

    def insert_item(self, obj):
        #・新しいデータ項目を生成
        node = self._Node(obj,None)
        #・チェーンが空の場合、チェーンの先頭に挿入
        if self.is_empty():
            self._head = node
        #・空でない場合、適切な箇所に挿入
        else:
            n = self._head
            if n._item > obj:
                node._next = n
                self._head = node
                return
            while n._next != None and n._next._item < obj:
                n = n._next
            node._next = n._next
            n._next = node
        #・データ数のインクリメント
        self._size += 1
    def delete_item(self):
        #・チェーンが空の場合、エラー
        if self.is_empty():
            print("Error")
            return
        #・チェーンが空でない場合、先頭データを返却
        res = self._head._item
        #・先頭データを削除
        self._head = self._head._next
        #・データ数のデクリメント
        self._size -= 1
        return res
    def show_list(self):
        n = self._head
        print ("List:", end="")
        while n != None :
            print (n._item, end=" ")
            n = n._next
        print ()
if __name__ == '__main__' :
    l = sort_list()
    l.insert_item(8)
    l.insert_item(5)
    l.show_list()
    l.insert_item(3)
    l.insert_item(10)
    l.insert_item(6)
    l.show_list()
    l.delete_item()
    l.show_list()
```

問題２
二分木にあるデータの中から最大値および最小値を求めるプログラムを
作成し、実行結果を示しなさい。
与えられたキーをもつデータ項目を探索し、見付かった場合は、そのデー
タを表示し、見付からなかった場合は、「Data not found」を表示するプロ
グラムを作成し、実行結果を示しなさい。

```python
class BinaryTree:

    class _Node:
        def __init__(self, item):
            self.item = item
            self.left = None
            self.right = None

    def __init__(self):
        self.root = None  
        self.size = 0

    def __len__(self):
        return self.size

    def insert(self, item):
        new = self._Node(item)
        if self.root is None:
            self.root = new
            self.size = 1  # サイズを1に設定
            return  # 処理終了
        cur = self.root
        Found = False
        while not Found:
            par = cur
            # itemが現在のノードより小さい場合、左へ
            if item < cur.item:
                cur = cur.left
                if cur is None: 
                    par.left = new
                    Found = True

            else:
                cur = cur.right
                if cur is None:  
                    par.right = new
                    Found = True
        self.size += 1
    def find_min(self):
        if self.root is None:
            return None # 木が空
        
        cur = self.root
        # 左の子が存在する限り、左へ移動し続ける
        while cur.left is not None:
            cur = cur.left
        return cur.item

    def find_max(self):
        if self.root is None:
            return None # 木が空

        cur = self.root
        # 右の子が存在する限り、右へ移動し続ける
        while cur.right is not None:
            cur = cur.right
        return cur.item

    def search(self, key):
        cur = self.root
        
        while cur is not None:
            if key == cur.item:
                return cur.item  # キーが見つかった
            elif key < cur.item:
                cur = cur.left   # キーが現在地より小さい場合、左へ
            else:
                cur = cur.right  # キーが現在地より大きい場合、右へ
                
        return None # ループを抜けた = 見つからなかった
    
tree = BinaryTree()
tree.insert(1)
tree.insert(5)
tree.insert(7)
tree.insert(2)
tree.insert(19)

print(tree.find_max())
print(tree.find_min())
if tree.search(5) != None:
    print("tree contains 5")
else:
    print("Data not found")
if tree.search(3) != None:
    print("tree contains 5")
else:
    print("Data not found")
```
