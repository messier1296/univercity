端末から入力された数の平均値および分散を
計算する関数を作成しなさい。入力は8.1, 7.9,
9.5, 8.8, 9.1のとき、その平均値および分散を
計算しなさい。

```python
def mean_and_var():
    # コンマ区切りで入力された数値をfloatに変換してリスト化
    lis = list(map(float, input().split(',')))

    # --- 平均の計算 ---
    ave = 0.0
    for i in lis:
        ave += i   # 各要素を合計
    ave /= len(lis)  # 合計を要素数で割って平均を求める

    # --- 分散の計算 ---
    var = 0.0
    for i in lis:
        var += (ave - i) ** 2  # 各値と平均の差の2乗を足し合わせる
    var /= len(lis)  # 全体の平均を取って分散を求める（母分散）

    # 結果を表示
    print("Mean and Var:{},{}".format(ave, var))
```

```python
mean_and_var()
```

成績からGPAを計算するプログラムを作成しなさい。 なお、辞書型データを使用しなさい。 score = {'A+': 4.0, 'A': 3.0, 'B': 2.0, 'C': 1.0, 'D': 0.0}

```python
def cal_gpa():
    # 成績と対応するGPA値の辞書を定義
    score = {'A+': 4.0, 'A': 3.0, 'B': 2.0, 'C': 1.0, 'D': 0.0}
    # 入力された成績を数値に変換して格納するリスト
    grades = []

    # 成績の入力を繰り返し受け取るループ
    while True:
        g = input("Input your grade, or RETURN to finish: ")
        if g == "":  # 空入力（Enterのみ）で終了
            break
        if g in score:  # 有効な成績なら辞書を参照して追加
            grades.append(score[g])
        else:  # 無効な入力はエラー表示して終了
            print("Error: invalid grade input.")
            return

    # 入力が1つ以上あれば平均（GPA）を計算
    if grades:
        gpa = sum(grades) / len(grades)
        print(f"Your GPA is {gpa:.2f}")  # 小数点以下2桁で表示
    else:
        print("No grades entered.")  # 空のまま終了した場合の対応

# 関数を実行
cal_gpa()
```

データ数Nの配列aを降順に並べ替えるのに最大値選択法があり、
次のような手順になっている。このアルゴリズムに従って、右の成績
を降順に並べ替えるプログラムを作成しなさい。

```python
def choice_sort(x):
    # 選択ソート（第2要素を基準に降順に並べ替える）
    for i in range(len(x)):
        max_val = x[i]  # 現在の位置を仮の最大値とする
        k = i            # 最大値のインデックスを記録
        for j in range(i, len(x)):
            # 第2要素（スコア）がより大きいものを見つけたら更新
            if x[j][1] > max_val[1]:
                max_val = x[j]
                k = j
        # 最大値を現在の位置iと交換
        x[k] = x[i]
        x[i] = max_val
    return x

# (学籍番号, 成績) のタプルのリスト
seiseki = [
    (100, 70),
    (101, 90),
    (102, 85),
    (103, 80),
    (104, 100),
    (105, 87)
]

# 成績の高い順に並べ替え
sorted_lis = choice_sort(seiseki)

# 結果を表示
print(sorted_lis)
```

下記‘-’の行を書く関数を再帰呼び出すことにより、
「実行結果」のように出力する関数draw()を作成し
なさい。

```python
def draw_line(length):
    # 指定された長さの「-」を出力する
    line = '-' * length
    print(line)

def draw(x, y):
    # 再帰の終了条件：x が 1 未満になったら終了
    if x < 1:
        return

    # y+1 から 1 まで、1ずつ減らしながら線を出力
    for i in range(y + 1, 0, -1):
        draw_line(i)

    # x を 1 減らして再帰呼び出し
    draw(x - 1, y)

# 関数呼び出し：最初に (x=2, y=5)
draw(2, 5)
```
