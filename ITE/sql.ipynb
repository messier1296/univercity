```python
def tables():
    tables = %sql select tbl_name from sqlite_master where type="table" or type="view"
    return list(tables)
```

```python
def schema(tbl_name="*"):
    schema = %sql select sql from sqlite_master where type="table" and tbl_name="{tbl_name}"
    return schema
```

```python
%%sql
drop table if exists 商品; /*すでに商品テーブルがあれば削除*/
create table 商品(
    商品番号 char(3) not null,
    商品名 varchar(20),
    価格 int,
    primary key(商品番号)
);
```

```python
%%sql
drop table if exists 顧客; /*すでに顧客テーブルがあれば削除*/
create table 顧客(
    顧客番号 int not null,
    顧客名 varchar(20),
    住所 varchar(100),
    primary key(顧客番号)
);
```
